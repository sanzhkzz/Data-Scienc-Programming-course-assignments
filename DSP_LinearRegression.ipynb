# 1.1 . Read the data

```python
X=[231,229.9,240.9,247.5,255.2,243.1,242,256.3,220,236.5,225.5,249.7]
Y=[226.6,220,224.4,236.5,244.2,237.6,239.8,221.1,233.2,224.4,224.4,233.2]
```

```python
l=len(X)
sum_x=sum(X)
sum_y=sum(Y)
x_y = [a * b for a, b in zip(X, Y)]
sum_x_y=sum(x_y)
x2 = [a ** 2 for a in X]
y2 =[a ** 2 for a in Y]
x2_sum=sum(x2)
y2_sum=sum(y2)
```

# 1.2 . Correlation

```python
corr_coef=(l*sum_x_y-(sum_x*sum_y))/math.sqrt((l*x2_sum-(sum(X)**2))*(l*y2_sum-(sum(Y)**2)))
corr_coef
```

# 1.3. Best-fitting straight line

```python
slope = (sum_x_y - sum_x*sum_y/l)/(x2_sum - sum(X)*sum(X)/l)
print(slope)
```

```python
intercept = sum(Y)/l - slope*sum(X)/l
print(intercept)
```

```python
X1 = np.array(X)
y2 = np.array(Y)
y_pred = intercept + slope*X1
print('y_pred:', y_pred)
```

# 1.4 Plotting

```python
plt.figure(figsize=(6,5))
plt.scatter(X1, y2, c='black', marker='x') 
plt.plot([min(X1), max(X1)], [min(y_pred), max(y_pred)], color='black')  # regression line
plt.title('Scatter diagram of sales ')
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['carat'], df['price'], test_size=0.2, random_state = 12)
X_train = np.array(X_train).reshape((-1, 1))
X_test = np.array(X_test).reshape((-1, 1))
```

# 3.2 Fit model

```python
model_lr = LinearRegression().fit(X_train, y_train)
train_pred = model_lr.predict(X_train)
test_pred = model_lr.predict(X_test)
```

# 3.3 Predicted responses

```python
print(test_pred)
```

# 3.4 Evaluating metrics for train and test

**Creating function in order to obtain all the metrics using print_evaluate and putting values into**

```python
from sklearn import metrics
def print_evaluate(true, predicted):  
    r2_square = metrics.r2_score(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    mae = metrics.mean_absolute_error(true, predicted)
    print('R-squared', np.round(r2_square,3))
    print('MSE:', np.round(mse,3))
    print('RMSE:', np.round(rmse,3))
    print('MAE:', np.round(mae,3))
```

```python
print('Answers for train')
print()

print_evaluate(y_train, train_pred)
```

```python
print('Answers for test:')
print()
print_evaluate(y_test, test_pred)
```
